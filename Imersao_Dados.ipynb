# **AULA 4** - Prevendo notas com Machine Learning

Nessa aula veremos um modelo de Machine Learning como prever a nota da prova

```python
#aqui separamos as notas que vamos nos basear
provas_entrada = ["NU_NOTA_CH", "NU_NOTA_LC", "NU_NOTA_CN", "NU_NOTA_REDACAO"]
#essa é a previsão de nota de matemática a partir das notas acima
prova_saida = "NU_NOTA_MT"
#farei esse processo para quando eu for rodar o LineatSVM ele não dar erros com os "NaN"s
dados_sem_notas_zero = dados_sem_notas_zero[provas].dropna()                
notas_entrada = dados_sem_notas_zero[provas_entrada]
notas_saida = dados_sem_notas_zero[prova_saida]
```

```python
notas_entrada
```

O que vemos aqui é muito parecido com a função f(x) = y, onde temos as notas de saída (y) em função das notas de entrada (x). Então vamos chamar as características que a gente tem de X e as notas que queremos prever de Y


```python
x = notas_entrada
y = notas_saida
```

Aqui não se tem ideia de como modelo tá performando esses dados pra criar o modelo, então é necessário separar os dados para treinar e outra pra testar esse modelo. Pra separar isso posso fazer de várias maneiras, mas existe uma biblioteca que facilita muito esse processo, o *scikit learning*. Ele traz modelos de classificação, regressão e clusterização, além de todo pré processamento do modelo de machine learning

```python
#importar dentro do model selection o método train test split
from sklearn.model_selection import train_test_split
#recebe os dados de entrada e o que eu quero de saída
train_test_split(x, y)
```

Esses dados que ele está plotando é o meu x de treino, x de teste, y de treino e y de teste. Isso tudo tem detalhado na documentação.

```python
#agora vou passar um outro paramêtro aqui q é a quantidade de teste. Nesse caso 25% dos dados para teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.25)
x_treino.head()
```

Nesse caso, sempre que eu for rodar a célula ele vai me dar os casos aleatóriamente. Para que ele sempre me de os mesmos dados então tenho que setar o random seed.

```python
#Como eu vou usar o seed pra outras funções vou criar uma váriavel 'SEED'. As linhas sempre vão ser as mesmas
SEED = 4321
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.25, random_state=SEED)
```

```python
#faz a predição do modelo. No parâmetro coloco as caracteristicas e então vai retornar a predição de notas em matemática e tecnologias em um array
modelo.predict(x_teste)
```

Quando vou verificar os erros não posso misturar as diferenças que deu positivas com as negativas. Porque, por exemplo, qdo eu erro 100 numa nota, e depois zero na outra e menos 100 em outra, ao somar esses valores vai dar como resultado zero. E isso é um erro gravíssimo porque posso estar pensando que meu modelo não tem erro nenhum. Por isso quando eu elevo esses valores ao quadrado eu evito esse tipo de erro. Isso se deve ao fato de eu ter poucas informações. Dependendo da minha base de dados, se eu tiver mais informações, o negativo e positivo podem se tornar fatores importantes. Entretando devemos tomar muito cuidado com o viés da predição, pois depende muito do contexto e essas análises devem ser feitas com acuidade.

```python
#rodando novamente agora elevando ao quadrado
resultados = pd.DataFrame()
resultados["Real"] = y_teste
resultados["Previsao"] = predicoes_matematica
resultados["diferenca"] = resultados["Real"] - resultados["Previsao"]
#criando outra coluna 
resultados["quadrado_diferenca"] = (resultados["Real"] - resultados["Previsao"])**2
resultados
```

### Árvore de decisão

```python
#arvore de decisão
from sklearn.tree import DecisionTreeRegressor

#divisão de dados de treino e de teste com um tamanho de 25
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.25)
#criando modelo
modelo_arvore = DecisionTreeRegressor(max_depth = 3)
modelo_arvore.fit(x_treino, y_treino)
predicoes_matematica_arvore = modelo_arvore.predict(x_teste) #fazendo a predição com o modelo arvore
#faz avaliação do treino
mean_squared_error(y_teste, predicoes_matematica_arvore)
```

Aqui ele sempre vai retornar um valor aleatório e dessa forma não sei qual valor é o correto. Mesmo que pegue um modelo com uma influência menor na aleatoriedade, no caso da árvore de decisão, ainda temos essa grande influência na aleatoriedade da forma que a árvore divide a nossa base de dados entre base de treino e base de teste. Dessa forma acaba somando dois fatores de aletoriedade no resultado. Então temos que reduzir essa aleatoriedade.

Essa amostra está separa em 4 partes. Isso significa que essa amostra vai ser testada 4 vezes... cada linha de testagem vai pegar 1 bloco pra testar (25%) e o restante vai ser pra treino. E assim sucessivamente. E então eu faço uma média desses teste para obter um resultado. Como indicado na figura: [fonte](https://scikit-learn.org/stable/_images/grid_search_cross_validation.png)
transferir.png

```python
#implementar o modelo cross validation no modelo decision tree
from sklearn.model_selection import cross_validate

modelo_arvore = DecisionTreeRegressor(max_depth=2)
#passa o modelo, features q é o x e os resultados q é o y
cross_validate(modelo_arvore, x, y)
```

Por padrão são 5 folds. Só que esses resultados estão bem diferentes do que estávamos trabalhando anteriormente. Por padrão o cross validation tem um parâmetro chamado *score*. E como estávamos trabalhando com *mean square error* e aqui, por padrão, ele seta um metódo chamado R2. Entao temos q setar esse método corretamente


```python
modelo_arvore = DecisionTreeRegressor(max_depth=2)
cross_validate(modelo_arvore, x, y, cv=10, scoring="neg_mean_squared_error") 
#aqui eu preciso colocoar o "neg" antes pq esse é o método de pontuação q no caso qto maior melhor. é uma regra de implementação. Por isso ele acaba ficando negativo
#então aqui ele fica negativo pra dizer q qto mais próximo de zero melhor.
#cv eu seto a qtde de splits (folds)
```

Quando se compara modelos é importante ver que além do score, que me diz o erro, é importante saber também o tempo de resposta.
Pra esse modelo que estamos trabalhando não interessa esse parâmetro. Mas se formos avaliar o tempo de resposta, esse parâmetro é importante. Pra uma base de dados temos que ver se vale a pena processar por muito tempo esses dados e se o processador aguenta. Outro exemplo mais prático seria no caso do Netflix, quando ele me dá um cardápio de filmes de acordo com minhas preferências eu preciso q esse tempo seja curto.

```python
#Então vamos transformar esse números em positivos e então nesse caso fica quanto menor melhor.
resultados = cross_validate(modelo_arvore, x, y, cv=10, scoring="neg_mean_squared_error")
resultados["test_score"]*-1
```

```python
#média dos resultados
media = (resultados["test_score"]*-1).mean()
media
```

Aqui temos um problema ao calcular a média. Eu posso estar pegando um resultado enviesado devido aos outliers. Por exemplo, quando eu tenho valores muito distantes entre si como 1 e 9, qdo faço a média acaba ficando 5, sendo q não tenho nenhum valor próximo a esse nos meus dados. Pra isso podemos calcular o desvio padrão e calcular o intervalo de confiança.

```python
modelo_arvore = DecisionTreeRegressor(max_depth=2)
resultados = cross_validate(modelo_arvore, x, y, cv= 10, scoring="neg_mean_squared_error")
media = (resultados["test_score"]*-1).mean()
#desvio padrão
desvio_padrao = (resultados["test_score"]*-1).std()
#para saber "as pontas" do meu desvio padrão eu acho os limites inferior e superior
lim_inferior = media - (2*desvio_padrao)
lim_superior = media + (2*desvio_padrao)

print(f"Intervalo de confiança {lim_inferior} - {lim_superior}")
```

95% das nossas previsões devem estar nesse intervalo de 6015 e 7100. É importante ressoaltar q esses resultados não são precisos nessa afirmação. Se eu pegar esse método e rodas 100 vezes eu vou 95% das vezes o número real, o score real. (ver melhor a definição)

```python
#criando uma função
def calcula_mse(resultados):
    media = (resultados["test_score"]*-1).mean()
    desvio_padrao = (resultados["test_score"]*-1).std()
    lim_inferior = media - (2*desvio_padrao)
    lim_superior = media + (2*desvio_padrao)
    print(f"Intervalo de confiança {lim_inferior} - {lim_superior}")
```
